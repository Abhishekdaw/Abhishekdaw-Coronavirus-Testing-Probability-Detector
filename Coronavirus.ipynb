# Train Test splitting

```python
import numpy as np
```

```python
def data_split(data,ratio):
    np.random.seed(42)
    shuffled=np.random.permutation(len(data))
    test_setsize=int(len(data)*ratio)
    test_indices=shuffled[:test_setsize]
    train_indices=shuffled[test_setsize:]
    return data.iloc[test_indices],data.iloc[train_indices]
    
```

```python
train,test=data_split(df,0.3)
```

```python
train
train.columns=train.columns.to_series().apply(lambda x: x.strip())
```

```python
test
test.columns=test.columns.to_series().apply(lambda x: x.strip())
```
